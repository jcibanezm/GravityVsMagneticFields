## Voronoi tesselation

---
This notebook intends to take an array of particle potisions and do a Voronoi (or maybe a Delaunay) Tesselation in order to recover the volume "encompassed" (that corresponds to) that particle. With the volume of the particle I can compute the mass within the voronoi cell and use that to compute the flux of kinetic energy accross the surface of the cloud.

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.spatial import Voronoi, voronoi_plot_2d

points = np.random.rand(20,2)
```

```python
vor = Voronoi(points)
```

```python
# Voronoi points

X_p = []
Y_p = []
for point in vor.points:
    X_p.append(point[0])
    Y_p.append(point[1])
X_p = np.array(X_p)
Y_p = np.array(Y_p)
```

```python
# Voronoi vertices

X_v = []
Y_v = []
for point in vor.vertices:
    X_v.append(point[0])
    Y_v.append(point[1])
X_v = np.array(X_v)
Y_v = np.array(Y_v)
```

```python
# Voronoi ridges betwen points

Vx_ridge = []
Vy_ridge = []

for pair in vor.ridge_points:
    Vx_ridge.append(0)
    Vy_ridge.append(0)
    
    Vx_ridge[-1] = []
    Vy_ridge[-1] = []
    
    Vx_ridge[-1].append(vor.points[pair[0]][0]) 
    Vy_ridge[-1].append(vor.points[pair[0]][1]) 

    Vx_ridge[-1].append(vor.points[pair[1]][0]) 
    Vy_ridge[-1].append(vor.points[pair[1]][1]) 

Vx_ridge = np.array(Vx_ridge)
Vy_ridge = np.array(Vy_ridge)
```

```python
# Voronoi ridges between vertices

Vx_vert = []
Vy_vert = []

for pair in vor.ridge_vertices:
    if not -1 in pair:
        Vx_vert.append(0)
        Vy_vert.append(0)

        Vx_vert[-1] = []
        Vy_vert[-1] = []

        Vx_vert[-1].append(vor.vertices[pair[0]][0]) 
        Vy_vert[-1].append(vor.vertices[pair[0]][1]) 

        Vx_vert[-1].append(vor.vertices[pair[1]][0]) 
        Vy_vert[-1].append(vor.vertices[pair[1]][1]) 

Vx_vert = np.array(Vx_vert)
Vy_vert = np.array(Vy_vert)
```

```python
# Get the number of closed Voronoi cells here.
vor_cells = 0
for ii in range(len(vor.regions)):
    if not -1 in vor.regions[ii]:
        vor_cells +=1 

Vor_x = []
Vor_y = []
for ii in range(len(vor.regions)):

    if not -1 in vor.regions[ii]:
        Fx = []
        Fy = []

        if not -1 in vor.regions[ii]:
            for i in range(len(vor.regions[ii])):

                if i < len(vor.regions[ii])-1:
                    ver0 = vor.regions[ii][i]
                    ver1 = vor.regions[ii][i+1]
                else:
                    ver0 = vor.regions[ii][i]
                    ver1 = vor.regions[ii][0]

                Fx.append(0)
                Fy.append(0)

                Fx[-1] = []
                Fy[-1] = []

                Fx[-1].append(vor.vertices[ver0][0])
                Fx[-1].append(vor.vertices[ver1][0])

                Fy[-1].append(vor.vertices[ver0][1])
                Fy[-1].append(vor.vertices[ver1][1])

        Fx = np.array(Fx)
        Fy = np.array(Fy)

        Vor_x.append(Fx)
        Vor_y.append(Fy)
    
Vor_x = np.array(Vor_x)
Vor_y = np.array(Vor_y)
```

```python
Area = []
for j in range(vor_cells):
    num_ver = len(Vor_x[j])
    Area_temp = 0
    for i in range(num_ver-1):
        Area_temp += Vor_x[j][i][0]*Vor_y[j][i][1] - Vor_x[j][i][1]*Vor_y[j][i][0]  
    Area.append(abs(Area_temp) * 0.5 )
Area = np.array(Area)
```

```python
min_A = min(Area)
max_A = max(Area)
dA    = (max_A - min_A)

Area_norm = (Area - min_A) / dA
```

```python
from scipy.spatial import Voronoi, Delaunay
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tetravol(a,b,c,d):
    """
    Calculates the volume of a tetrahedron, given vertices a,b,c and d.
    """
    tetravol = abs(np.dot((a-d), np.cross((b-d), (c-d))))/6
    return tetravol
```

```python
def vol(vor, p):
    """ 
    Calculate the volume of a 3D Voronoi cell based on point p and the voronoi diagram.
    """
    dpoints = []
    vol     = 0
    
    # Loop over the number of vertices in a given region.
    for v in vor.regions[vor.point_region[p]]:
        
        # Append the vector of each vertice to dpoints array.
        dpoints.append(list(vor.vertices[v]))
        
    # Perform a Delaunay tessalation for the polyhedron given by all the vertices in dpoints.
    tri=Delaunay(np.array(dpoints))
    
    # Loop over the various tetrahedrons returned by the Delaunay tessalation.
    for simplex in tri.simplices:
        
        # Compute the volume of the tetrahedron.
        vol += tetravol(np.array(dpoints[simplex[0]]),np.array(dpoints[simplex[1]]),
                        np.array(dpoints[simplex[2]]),np.array(dpoints[simplex[3]]))
    
    # Return the total volume of the polyhedron (sum of the volume of all the polyhedrons) 
    return vol
```

```python
for i,p in enumerate(vor.points):
    pvol = 0
    out = False
    for v in vor.regions[vor.point_region[i]]:
        if v<=-1:
            out=True
        else:
            if not out:
                pvol = vol(vor, i)
    print("Voronoi pint", i,"with coordinates", p,"Has a volume of", pvol)
```
