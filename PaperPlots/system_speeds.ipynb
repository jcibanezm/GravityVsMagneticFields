```python
import yt
import os
import math
from yt import derived_field
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from yt.units import pc, kpc, second, Kelvin, gram, erg, cm

import Read_binary as Rd
import Read_Crutcher_data as ReadB
import Reconstruct_UG as RUG

import cPickle as cP

%matplotlib inline

Laptop = False

# Define some constant parameters to be used.
mp      = 1.6726e-24  * gram # g
mu      = 1.2924
kb      = 1.3806e-16  *erg / Kelvin # erg K-1
GNewton = 6.6743e-8   * cm**3 / (gram * second**2 )# cm3 g-1 s-2
Msun    = 1.9884e33   * gram
mm      = mu*mp

# Create a derived field.
@derived_field(name="numdens", units="1/cm**3", force_override=True)
def numdens(field, data):
    #mp      = 1.6726e-24 * gram
    #mu      = 1.2924
    #mm      = mu*mp
    dens_here = data["dens"].value
    dens_here = dens_here * gram / cm**3
    return dens_here/mm

yt.add_field('numdens', function=numdens, units="1/cm**3", force_override=True)
```

```python
# Logaritmic bins.

nbins = 50
nbined = np.logspace(-2, 4.3, num=nbins)
```

```python
# Calculate the total energy density for each density bin

alfven   = []
alfvenMW = []
alfvenVW = []
cs       = []
cs_MW    = []
cs_VW    = []
beta     = []
sigma    = []
sigma_MW = []
sigma_VW = []

delta_sigma  = []
delta_alfven = []
delta_cs     = []

delta_sigma_VW  = []
delta_alfven_VW = []
delta_cs_VW     = []

Mach_mag       = []
delta_Mach_Mag = []

mass_bin = []
vol_bin  = []
nbins    = []
valf_MW  = []
valf_VW  = []

#-----------------------------------------------------------------------
# P = nkT
pressure = np.array(my_box["numdens"] * kb * my_box["temp"])

#-----------------------------------------------------------------------
# Bmag = B^2
Bmag = np.sqrt( my_box["magx"]* my_box["magx"] + 
                my_box["magy"]* my_box["magy"] + 
                my_box["magz"]* my_box["magz"]) 

valf = np.sqrt(Bmag*Bmag / (4.*math.pi*my_box["numdens"]*mm))


#-----------------------------------------------------------------------
# vrel = v_i - BV
vrelx = np.array(my_box["velx"] - cloud_props["BV_x"])
vrely = np.array(my_box["vely"] - cloud_props["BV_y"])
vrelz = np.array(my_box["velz"] - cloud_props["BV_z"])

# velocity magnitude
# vmag = (vrel^2)^0.5
vmag    = np.sqrt(vrelx**2 + vrely**2 + vrelz**2)

# Plasma Beta parameter
Beta = (8.*math.pi *pressure / (Bmag*Bmag))

Mach_all = vmag / valf

npmean = True

for bbin in range(len(nbined)):
    if bbin == 0:        
        nlow      = nbined[bbin] - (nbined[bbin+1] - nbined[bbin])/2.0
        nup       = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
        
    elif bbin != len(nbined)-1:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup  = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
    
    else:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup = nbined[bbin] + (nbined[bbin] - nbined[bbin-1])/2.0        
        
    # True if the cell is within the appropriate density range.
    within_range = (my_box["numdens"]>=nlow) & (my_box["numdens"]<nup)
    zeros        = np.where(within_range == False)
        
    # Compute the number of bins here.
    bins_here    = np.delete(my_box["cell_mass"], zeros)
    nbins.append(len(bins_here))
        
    # Compute the total mass in this bin.
    Mass_here_arr    = np.delete(my_box["cell_mass"], zeros)
    Mass_here_tot    = np.sum(Mass_here_arr)
    mass_bin.append(Mass_here_tot)

    # Compute the total Volume in this bin.
    Vol_here_arr    = np.delete(my_box["cell_volume"], zeros)
    Vol_here_tot    = np.sum(Vol_here_arr)
    vol_bin.append(Vol_here_tot)
    
    # Compute the mean sound speed.
    cs_here     = np.sqrt(4./3. * kb * my_box["temp"]/mm ) 
    cs_here_arr = np.delete(cs_here, zeros)
    if npmean:
        cs_here_tot = np.mean(cs_here)
    else:
        cs_here_tot = np.sum(cs_here) / len(cs_here)            
    cs.append(cs_here_tot)    
    
    # Compute the mean sound speed, mass weighted
    cs_arr = np.delete(cs_here, zeros)        
    cs_MW_tot = np.sum(cs_arr * Mass_here_arr) / (Mass_here_tot)
    cs_MW.append(cs_MW_tot)
                
    # Compute the mean sound speed, Volume weighted
    cs_VW_tot = np.sum(cs_arr * Vol_here_arr) / (Vol_here_tot)
    cs_VW.append(cs_VW_tot)
        
    # Calculate the dispersion around the mass weighted mean sound speed.
    delta_cs_tot = np.sqrt(np.sum((cs_arr - cs_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_cs.append(delta_cs_tot)

    # Calculate the dispersion around the mass weighted mean sound speed.
    delta_cs_tot_VW = np.sqrt(np.sum((cs_arr - cs_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_cs_VW.append(delta_cs_tot_VW)

    
    # Compute the mean turbulent velocity
    meanv = 0
    meanv = np.delete(vmag, zeros)
    if npmean:
        meanv = np.mean(meanv)
    else:
        meanv = np.sum(meanv) / len(meanv)        
    sigma.append(meanv)
        
    # Compute the mean turbulent velocity, mass weighted
    meanv_arr = np.delete(vmag, zeros)    
    
    meanv_MW_tot = np.sum(meanv_arr * Mass_here_arr) / (Mass_here_tot)
    sigma_MW.append(meanv_MW_tot)
                
    # Compute the mean turbulent velocity, volume weighted
    meanv_VW_tot = np.sum(meanv_arr * Vol_here_arr) / (Vol_here_tot)
    sigma_VW.append(meanv_VW_tot)
    
    # Calculate the dispersion around the turbulent rms.
    delta_sigma_tot = np.sqrt(np.sum((meanv_arr - sigma_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    #delta_sigma_tot = np.sqrt(np.sum((meanv_arr - sigma_VW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_sigma.append(delta_sigma_tot)

    # Calculate the dispersion around the turbulent rms.
    delta_sigma_tot_VW = np.sqrt(np.sum((meanv_arr - sigma_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_sigma_VW.append(delta_sigma_tot_VW)
    
    # Compute the mean alfven velocity
    valf_here = np.sqrt(Bmag*Bmag / (4.*math.pi*my_box["numdens"]*mm))
    valf_here = np.delete(valf_here, zeros)
    
    if npmean:
        valf_here = np.mean(valf_here)
    else:
        valf_here = np.sum(valf_here) / len(valf_here)            
    alfven.append(valf_here)
      
    # Compute the mass weighted and volume weighted alfven speed.
    valf_here_arr = np.delete(valf, zeros)
    
    valf_MW_tot = np.sum(valf_here_arr * Mass_here_arr) / Mass_here_tot
    valf_MW.append(valf_MW_tot)

    valf_VW_tot = np.sum(valf_here_arr * Vol_here_arr) / Vol_here_tot
    valf_VW.append(valf_VW_tot)
    
    # Calculate the dispersion aroung the alfven velocity.
    delta_alfven_tot = np.sqrt(np.sum((valf_here_arr - valf_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_alfven.append(delta_alfven_tot)
    
    # Calculate the dispersion aroung the alfven velocity.
    delta_alfven_tot_VW = np.sqrt(np.sum((valf_here_arr - valf_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_alfven_VW.append(delta_alfven_tot_VW)
    
    # Magnetic Mach number
    Mach_Mag_arr = np.delete(Mach_all, zeros)
    
    Mach_Mag_tot = np.sum(Mach_Mag_arr * Mass_here_arr) / Mass_here_tot
    Mach_mag.append(Mach_Mag_tot)

    # Calculate the dispersion aroung the alfven velocity.
    delta_Mach_Mag_tot = np.sqrt(np.sum((Mach_Mag_arr - Mach_mag[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_Mach_Mag.append(delta_Mach_Mag_tot)
 
    # mean Beta
    beta_here = 0
    beta_here = np.delete(Beta, zeros)
        
    if npmean:
        beta_here = np.mean(beta_here)
    else:
        beta_here = np.sum(beta_here) / len(beta_here)            
    beta.append(beta_here)
        
#------------------------------------------------------------------
alfven = np.array(alfven)
cs     = np.array(cs)
cs_MW  = np.array(cs_MW)
cs_VW  = np.array(cs_VW)
beta   = np.array(beta)
sigma  = np.array(sigma)
sigma_MW = np.array(sigma_MW)
sigma_VW = np.array(sigma_VW)
mass_bin = np.array(mass_bin)
vol_bin = np.array(vol_bin)
valf_MW  = np.array(valf_MW)
valf_VW  = np.array(valf_VW)

delta_sigma    = np.array(delta_sigma)
delta_alfven   = np.array(delta_alfven)
delta_Mach_Mag = np.array(delta_Mach_Mag)
delta_cs       = np.array(delta_cs)

delta_sigma_VW    = np.array(delta_sigma_VW)
delta_alfven_VW   = np.array(delta_alfven_VW)
delta_cs_VW       = np.array(delta_cs_VW)
```
