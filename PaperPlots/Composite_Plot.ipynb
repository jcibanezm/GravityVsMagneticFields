```python
# Restore the data in a cloud, compute the energy densities.
Cloud_name = "M4e3"
resolution = 0.5

snapshot   = 20
```

```python
import yt
import os
import math
from yt import derived_field
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from yt.units import pc, kpc, second, Kelvin, gram, erg, cm

import Read_binary as Rd
import Read_Crutcher_data as ReadB
import Reconstruct_UG as RUG

import cPickle as cP

%matplotlib inline

Laptop = False

# Define some constant parameters to be used.
mp      = 1.6726e-24  * gram # g
mu      = 1.2924
kb      = 1.3806e-16  *erg / Kelvin # erg K-1
GNewton = 6.6743e-8   * cm**3 / (gram * second**2 )# cm3 g-1 s-2
Msun    = 1.9884e33   * gram
mm      = mu*mp

# Create a derived field.
@derived_field(name="numdens", units="1/cm**3", force_override=True)
def numdens(field, data):
    #mp      = 1.6726e-24 * gram
    #mu      = 1.2924
    #mm      = mu*mp
    dens_here = data["dens"].value
    dens_here = dens_here * gram / cm**3
    return dens_here/mm

yt.add_field('numdens', function=numdens, units="1/cm**3", force_override=True)
```

```python
if resolution < 0.1:
    resolution_str = "%.3i" %(resolution*100)
else:
    resolution_str = "%.2i" %(resolution*10)
```

```python
#data_dir = "/data/manda/jcibanezm/StratBox/Data_In_gamera/CloudVicinity/%s/%spc" %(Cloud_name, resolution_str)
#data_dir = "/data/gamera/jcibanezm/StratBox/AccretionPaper/KineticEnergy/Cloud_Object/Binary%s/%spc" %(Cloud_name, resolution_str)
data_dir = "/data/gamera/jcibanezm/StratBox/AccretionPaper/KineticEnergy/CloudVicinity/M4e3/05pc"
```

```python
my_box                  = Rd.Read_Box_Snapshot(Cloud_name, resolution, snapshot, dir=data_dir)
#my_box                  = Rd.Read_Box_Snapshot(Cloud_name, resolution, snapshot)
cloud_data, cloud_props = Rd.Restore_One_Snapshot(Cloud_name, resolution, snapshot)
```

```python
# Calculate the total energy density for each density bin

# Logaritmic bins.

nbins = 50
nbined = np.logspace(-2, 4.3, num=nbins)

alfven   = []
alfvenMW = []
alfvenVW = []
cs       = []
cs_MW    = []
cs_VW    = []
beta     = []
sigma    = []
sigma_MW = []
sigma_VW = []

delta_sigma  = []
delta_alfven = []
delta_cs     = []

delta_sigma_VW  = []
delta_alfven_VW = []
delta_cs_VW     = []

Mach_mag       = []
delta_Mach_Mag = []

mass_bin = []
vol_bin  = []
nbins    = []
valf_MW  = []
valf_VW  = []

#-----------------------------------------------------------------------
# P = nkT
pressure = np.array(my_box["numdens"] * kb * my_box["temp"])

#-----------------------------------------------------------------------
# Bmag = B^2
Bmag = np.sqrt( my_box["magx"]* my_box["magx"] + 
                my_box["magy"]* my_box["magy"] + 
                my_box["magz"]* my_box["magz"]) 

valf = np.sqrt(Bmag*Bmag / (4.*math.pi*my_box["numdens"]*mm))


#-----------------------------------------------------------------------
# vrel = v_i - BV
vrelx = np.array(my_box["velx"] - cloud_props["BV_x"])
vrely = np.array(my_box["vely"] - cloud_props["BV_y"])
vrelz = np.array(my_box["velz"] - cloud_props["BV_z"])

# velocity magnitude
# vmag = (vrel^2)^0.5
vmag    = np.sqrt(vrelx**2 + vrely**2 + vrelz**2)

# Plasma Beta parameter
Beta = (8.*math.pi *pressure / (Bmag*Bmag))

Mach_all = vmag / valf

npmean = True

for bbin in range(len(nbined)):
    if bbin == 0:        
        nlow      = nbined[bbin] - (nbined[bbin+1] - nbined[bbin])/2.0
        nup       = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
        
    elif bbin != len(nbined)-1:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup  = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
    
    else:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup = nbined[bbin] + (nbined[bbin] - nbined[bbin-1])/2.0        
        
    # True if the cell is within the appropriate density range.
    within_range = (my_box["numdens"]>=nlow) & (my_box["numdens"]<nup)
    zeros        = np.where(within_range == False)
        
    # Compute the number of bins here.
    bins_here    = np.delete(my_box["cell_mass"], zeros)
    nbins.append(len(bins_here))
        
    # Compute the total mass in this bin.
    Mass_here_arr    = np.delete(my_box["cell_mass"], zeros)
    Mass_here_tot    = np.sum(Mass_here_arr)
    mass_bin.append(Mass_here_tot)

    # Compute the total Volume in this bin.
    Vol_here_arr    = np.delete(my_box["cell_volume"], zeros)
    Vol_here_tot    = np.sum(Vol_here_arr)
    vol_bin.append(Vol_here_tot)
    
    # Compute the mean sound speed.
    cs_here     = np.sqrt(4./3. * kb * my_box["temp"]/mm ) 
    cs_here_arr = np.delete(cs_here, zeros)
    if npmean:
        cs_here_tot = np.mean(cs_here)
    else:
        cs_here_tot = np.sum(cs_here) / len(cs_here)            
    cs.append(cs_here_tot)    
    
    # Compute the mean sound speed, mass weighted
    cs_arr = np.delete(cs_here, zeros)        
    cs_MW_tot = np.sum(cs_arr * Mass_here_arr) / (Mass_here_tot)
    cs_MW.append(cs_MW_tot)
                
    # Compute the mean sound speed, Volume weighted
    cs_VW_tot = np.sum(cs_arr * Vol_here_arr) / (Vol_here_tot)
    cs_VW.append(cs_VW_tot)
        
    # Calculate the dispersion around the mass weighted mean sound speed.
    delta_cs_tot = np.sqrt(np.sum((cs_arr - cs_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_cs.append(delta_cs_tot)

    # Calculate the dispersion around the mass weighted mean sound speed.
    delta_cs_tot_VW = np.sqrt(np.sum((cs_arr - cs_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_cs_VW.append(delta_cs_tot_VW)

    
    # Compute the mean turbulent velocity
    meanv = 0
    meanv = np.delete(vmag, zeros)
    if npmean:
        meanv = np.mean(meanv)
    else:
        meanv = np.sum(meanv) / len(meanv)        
    sigma.append(meanv)
        
    # Compute the mean turbulent velocity, mass weighted
    meanv_arr = np.delete(vmag, zeros)    
    
    meanv_MW_tot = np.sum(meanv_arr * Mass_here_arr) / (Mass_here_tot)
    sigma_MW.append(meanv_MW_tot)
                
    # Compute the mean turbulent velocity, volume weighted
    meanv_VW_tot = np.sum(meanv_arr * Vol_here_arr) / (Vol_here_tot)
    sigma_VW.append(meanv_VW_tot)
    
    # Calculate the dispersion around the turbulent rms.
    delta_sigma_tot = np.sqrt(np.sum((meanv_arr - sigma_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    #delta_sigma_tot = np.sqrt(np.sum((meanv_arr - sigma_VW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_sigma.append(delta_sigma_tot)

    # Calculate the dispersion around the turbulent rms.
    delta_sigma_tot_VW = np.sqrt(np.sum((meanv_arr - sigma_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_sigma_VW.append(delta_sigma_tot_VW)
    
    # Compute the mean alfven velocity
    valf_here = np.sqrt(Bmag*Bmag / (4.*math.pi*my_box["numdens"]*mm))
    valf_here = np.delete(valf_here, zeros)
    
    if npmean:
        valf_here = np.mean(valf_here)
    else:
        valf_here = np.sum(valf_here) / len(valf_here)            
    alfven.append(valf_here)
      
    # Compute the mass weighted and volume weighted alfven speed.
    valf_here_arr = np.delete(valf, zeros)
    
    valf_MW_tot = np.sum(valf_here_arr * Mass_here_arr) / Mass_here_tot
    valf_MW.append(valf_MW_tot)

    valf_VW_tot = np.sum(valf_here_arr * Vol_here_arr) / Vol_here_tot
    valf_VW.append(valf_VW_tot)
    
    # Calculate the dispersion aroung the alfven velocity.
    delta_alfven_tot = np.sqrt(np.sum((valf_here_arr - valf_MW[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_alfven.append(delta_alfven_tot)
    
    # Calculate the dispersion aroung the alfven velocity.
    delta_alfven_tot_VW = np.sqrt(np.sum((valf_here_arr - valf_VW[-1])**2 * Vol_here_arr) / (Vol_here_tot))
    delta_alfven_VW.append(delta_alfven_tot_VW)
    
    # Magnetic Mach number
    Mach_Mag_arr = np.delete(Mach_all, zeros)
    
    Mach_Mag_tot = np.sum(Mach_Mag_arr * Mass_here_arr) / Mass_here_tot
    Mach_mag.append(Mach_Mag_tot)

    # Calculate the dispersion aroung the alfven velocity.
    delta_Mach_Mag_tot = np.sqrt(np.sum((Mach_Mag_arr - Mach_mag[-1])**2 * Mass_here_arr) / (Mass_here_tot))
    delta_Mach_Mag.append(delta_Mach_Mag_tot)
 
    # mean Beta
    beta_here = 0
    beta_here = np.delete(Beta, zeros)
        
    if npmean:
        beta_here = np.mean(beta_here)
    else:
        beta_here = np.sum(beta_here) / len(beta_here)            
    beta.append(beta_here)
        
#------------------------------------------------------------------
alfven = np.array(alfven)
cs     = np.array(cs)
cs_MW  = np.array(cs_MW)
cs_VW  = np.array(cs_VW)
beta   = np.array(beta)
sigma  = np.array(sigma)
sigma_MW = np.array(sigma_MW)
sigma_VW = np.array(sigma_VW)
mass_bin = np.array(mass_bin)
vol_bin = np.array(vol_bin)
valf_MW  = np.array(valf_MW)
valf_VW  = np.array(valf_VW)

delta_sigma    = np.array(delta_sigma)
delta_alfven   = np.array(delta_alfven)
delta_Mach_Mag = np.array(delta_Mach_Mag)
delta_cs       = np.array(delta_cs)

delta_sigma_VW    = np.array(delta_sigma_VW)
delta_alfven_VW   = np.array(delta_alfven_VW)
delta_cs_VW       = np.array(delta_cs_VW)
```

```python
load_dir = "/data/gamera/jcibanezm/StratBox/MagneticCloudsPaper/PickleClouds/%s"%(Cloud_name)
filename = "Cloud_%s_snp%.3i_Forces.pkl"%(Cloud_name, snapshot)

f = open('%s/%s'%(load_dir, filename), 'rb') 
cloud_dict = cP.load(f)
f.close()
```

```python
# Calculate the magnitude of the forces.

P_grad_mag = np.sqrt(cloud_dict["Pgrad_x"]*cloud_dict["Pgrad_x"] + 
                     cloud_dict["Pgrad_y"]*cloud_dict["Pgrad_y"] + 
                     cloud_dict["Pgrad_z"]*cloud_dict["Pgrad_z"])

Lorentz_mag = np.sqrt(cloud_dict["lorentz_x"]*cloud_dict["lorentz_x"] +
                      cloud_dict["lorentz_y"]*cloud_dict["lorentz_y"] +
                      cloud_dict["lorentz_z"]*cloud_dict["lorentz_z"])

#Advection_mag = np.sqrt(cloud_dict["conv_x"]*cloud_dict["conv_x"] +
#                        cloud_dict["conv_y"]*cloud_dict["conv_y"] +
#                        cloud_dict["conv_z"]*cloud_dict["conv_z"])

Advection_mag = np.sqrt(cloud_dict["conv_x2"]*cloud_dict["conv_x2"] +
                        cloud_dict["conv_y2"]*cloud_dict["conv_y2"] +
                        cloud_dict["conv_z2"]*cloud_dict["conv_z2"])

Gacc_mag = np.sqrt(cloud_dict["gacc_x"]*cloud_dict["gacc_x"] +
                   cloud_dict["gacc_y"]*cloud_dict["gacc_y"] +
                   cloud_dict["gacc_z"]*cloud_dict["gacc_z"])

v2_grad_mag = np.sqrt(cloud_dict["ramp_x"]*cloud_dict["ramp_x"] +
                      cloud_dict["ramp_y"]*cloud_dict["ramp_y"] +
                      cloud_dict["ramp_z"]*cloud_dict["ramp_z"])

curl_vort_mag = np.sqrt(cloud_dict["curl_vort_x"]*cloud_dict["curl_vort_x"] +
                        cloud_dict["curl_vort_y"]*cloud_dict["curl_vort_y"] +
                        cloud_dict["curl_vort_z"]*cloud_dict["curl_vort_z"])

# Logaritmic bins.

numbins = nbins
max_dens = math.log10(max(cloud_dict["numdens"]))
#nbined = np.logspace(-2, max_dens, num=numbins)

Nbins    = np.array(np.zeros(len(nbined)))

mass_bin = np.zeros_like(Nbins)
vol_bin  = np.zeros_like(Nbins)

pgrad_acc_bin       = np.zeros_like(Nbins)
pgrad_acc_std_bin   = np.zeros_like(Nbins)

lorentz_acc_bin     = np.zeros_like(Nbins)
lorentz_acc_std_bin = np.zeros_like(Nbins)

advec_acc_bin       = np.zeros_like(Nbins)
advec_acc_std_bin   = np.zeros_like(Nbins)

gacc_bin            = np.zeros_like(Nbins)
gacc_std_bin        = np.zeros_like(Nbins)

ramp_bin            = np.zeros_like(Nbins)
ramp_std_bin        = np.zeros_like(Nbins)

curl_vort_bin            = np.zeros_like(Nbins)
curl_vort_std_bin        = np.zeros_like(Nbins)

numdens_mean_bin      = np.zeros_like(Nbins)

for bbin in range(numbins):
    if bbin == 0:        
        nlow      = nbined[bbin] - (nbined[bbin+1] - nbined[bbin])/2.0
        nup       = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
        
    elif bbin != len(nbined)-1:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup  = nbined[bbin] + (nbined[bbin+1] - nbined[bbin])/2.0
    
    else:
        nlow = nbined[bbin] - (nbined[bbin] - nbined[bbin-1])/2.0
        nup = nbined[bbin] + (nbined[bbin] - nbined[bbin-1])/2.0        
        
    # True if the cell is within the appropriate density range.
    within_range = (cloud_dict["numdens"]>=nlow) & (cloud_dict["numdens"]<nup)
    zeros        = np.where(within_range == False)
        
    # Compute the number of bins here.
    bins_here    = np.delete(cloud_dict["cell_mass"], zeros)
    Nbins[bbin]  = len(bins_here)
    
    # Make an array with ones everywhere.
    unity = bins_here / bins_here
    
    # get the density of the cells in this bin.
    dens_here_arr    = np.delete(cloud_dict["dens"], zeros)
    numdens_here_arr = np.delete(cloud_dict["numdens"], zeros)
    
    # mean number density of the cells.
    numdens_mean_bin[bbin] = np.average(numdens_here_arr)
    
    # Compute the total mass in this bin.
    Mass_here_arr    = np.delete(cloud_dict["cell_mass"], zeros)
    Mass_here_tot    = np.sum(Mass_here_arr)
    mass_bin[bbin]   = Mass_here_tot
    
    # Compute the total Volume in this bin.
    dx_here_arr     = np.delete(cloud_dict["dx"], zeros)
    Vol_here_tot    = np.sum(dx_here_arr*dx_here_arr*dx_here_arr)
    vol_bin[bbin]   = Vol_here_tot
    
    # Compute the gravitational acceleration.
    Pgrad_here_arr  = np.delete(P_grad_mag, zeros) 
    Pgrad_acc_arr   = Pgrad_here_arr / dens_here_arr
    Pgrad_acc_avg   = np.average(Pgrad_acc_arr)
    Pgrad_acc_std   = np.sqrt(np.average((Pgrad_acc_arr-Pgrad_acc_avg)**2, weights=unity))
    
    pgrad_acc_bin[bbin]     = Pgrad_acc_avg
    pgrad_acc_std_bin[bbin] = Pgrad_acc_avg
    
    del Pgrad_here_arr, Pgrad_acc_arr, Pgrad_acc_avg, Pgrad_acc_std
    
    # Calculate the accelerations from Lorentz forces.
    Lorentz_here_arr  = np.delete(Lorentz_mag, zeros) 
    Lorentz_acc_arr   = Lorentz_here_arr / dens_here_arr
    
    Lorentz_acc_avg   = np.average(Lorentz_acc_arr)
    Lorentz_acc_std   = np.sqrt(np.average((Lorentz_acc_arr-Lorentz_acc_avg)**2, weights=unity)) 
    
    lorentz_acc_bin[bbin]     = Lorentz_acc_avg
    lorentz_acc_std_bin[bbin] = Lorentz_acc_std
    
    del Lorentz_here_arr, Lorentz_acc_arr, Lorentz_acc_avg, Lorentz_acc_std
    
    # Calculate the accelerations from background advection
    Advec_here_arr  = np.delete(Advection_mag, zeros) 
    
    Advec_acc_avg   = np.average(Advec_here_arr)
    Advec_acc_std   = np.sqrt(np.average((Advec_here_arr-Advec_acc_avg)**2, weights=unity)) 
    
    advec_acc_bin[bbin]     = Advec_acc_avg
    advec_acc_std_bin[bbin] = Advec_acc_std
    
    del Advec_here_arr, Advec_acc_avg, Advec_acc_std
    
    # Calculate the accelerations from background advection
    Gacc_here_arr  = np.delete(Gacc_mag, zeros) 
    
    Gacc_acc_avg   = np.average(Gacc_here_arr)
    Gacc_acc_std   = np.sqrt(np.average((Gacc_here_arr-Gacc_acc_avg)**2, weights=unity)) 
    
    gacc_bin[bbin]     = Gacc_acc_avg
    gacc_std_bin[bbin] = Gacc_acc_std
    
    del Gacc_here_arr, Gacc_acc_avg, Gacc_acc_std
    
    # Calculate the accelerations from background advection
    RamP_here_arr  = np.delete(v2_grad_mag, zeros) 
    
    RamP_acc_avg   = np.average(RamP_here_arr)
    RamP_acc_std   = np.sqrt(np.average((RamP_here_arr-RamP_acc_avg)**2, weights=unity)) 
    
    ramp_bin[bbin]     = RamP_acc_avg
    ramp_std_bin[bbin] = RamP_acc_std
    
    del RamP_here_arr, RamP_acc_avg, RamP_acc_std
    
    
    # curl_vort_mag
    
    # Calculate the accelerations from background advection
    curlVort_here_arr  = np.delete(curl_vort_mag, zeros) 
    
    curlVort_acc_avg   = np.average(curlVort_here_arr)
    curlVort_acc_std   = np.sqrt(np.average((curlVort_here_arr-curlVort_acc_avg)**2, weights=unity)) 
    
    curl_vort_bin[bbin]     = curlVort_acc_avg
    curl_vort_std_bin[bbin] = curlVort_acc_std
    
    del curlVort_here_arr, curlVort_acc_avg, curlVort_acc_std
```
