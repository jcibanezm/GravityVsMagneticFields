```python
Cloud_name  = "M4e3"
resolution  = 0.06
ncut        = 100
snapshot    = 20

import yt
import os
import math
from yt import derived_field
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from yt.units import pc, kpc, second, Kelvin, gram, erg, cm
import cPickle as cP

import timeit

import scikits.vectorplot
import lic_internal

%matplotlib inline
```

```python
# Define some constant parameters to be used.
mp      = 1.6726e-24  * gram # g
mu      = 1.2924
kb      = 1.3806e-16  *erg / Kelvin # erg K-1
GNewton = 6.6743e-8   * cm**3 / (gram * second**2 )# cm3 g-1 s-2
Msun    = 1.9884e33   * gram
mm      = mu*mp

ppc = 3.0856776e18

# Create a derived field.
@derived_field(name="numdens", units="1/cm**3", force_override=True)
def numdens(field, data):
    #mp      = 1.6726e-24 * gram
    #mu      = 1.2924
    #mm      = mu*mp
    dens_here = data["dens"].value
    dens_here = dens_here * gram / cm**3
    return dens_here/mm

yt.add_field('numdens', function=numdens, units="1/cm**3", force_override=True)

# Generalize for all the other clouds.
if Cloud_name == "M4e3":

    px = 180
    py = -25
    pz =  10
    cloud_alpha = 0.4

    rad = 40
    real_rad = 30

    px_str = str(px)
    py_str = str(py)
    pz_str = str(pz)
    pz = pz - 3.

    # Mean bulk velocity of the cloud.
    bvx_mean = -1.
    bvy_mean = -1.
    bvz_mean = -1.

    txt_xoffset = 15
    txt_yoffset = -25


if Cloud_name == "M3e3":

    px = 450
    py = -380
    pz = 25
    cloud_alpha = 0.4

    rad = 60
    real_rad = 25

    px_str = str(px)
    py_str = str(py)
    pz_str = str(pz)

    px = px + 8
    py = py - 5
    pz = pz - 8

    # Mean bulk velocity of the cloud.
    bvx_mean = 0.
    bvy_mean = 3.0
    bvz_mean = -2.0

    txt_xoffset = 25
    txt_yoffset = -35
elif Cloud_name == "M8e3":
    px = 60
    py = 370
    pz = 30
    cloud_alpha = 0.8

    rad = 100
    real_rad = 30

    px_str = str(px)
    py_str = str(py)
    pz_str = str(pz)

    # Mean bulk velocity of the cloud.
    bvx_mean =  1.
    bvy_mean = -1.
    bvz_mean = -1.

    px = px + 5
    py = py - 10
    pz = pz - 5

    txt_xoffset = 30
    txt_yoffset = -50

cloud_dir = Cloud_name + "_" + "a%.2i" %(cloud_alpha*10) + "_x" + px_str + "_y" + py_str + "_z" + pz_str

if resolution < 0.1:
    resolution_str = "%.3i" %(resolution*100)
elif resolution == 1:
    resolution_str = "1"
else:
    resolution_str = "%.2i" %(resolution*10)
data_dir = "/data/manda/jcibanezm/StratBox/RealProd/1pc_and_AccClouds/AccPaper_Resims/Particles/" + cloud_dir + "/%spc/" %resolution_str

# Set the address and the name of the data set.
if resolution == 1:
    basename = "Strat_Box_" + Cloud_name + "_%spc_SG_hdf5_" %(resolution_str)
else:
    basename = "Strat_Box_" + Cloud_name + "_%spc_hdf5_" %(resolution_str)
```

```python
plt_file  = data_dir + basename + "plt_cnt_%.4i" %snapshot

# Load data set
pf = yt.load(plt_file)
```

```python
# Move the center of the frame.
px_now = px + bvx_mean * snapshot/10.
py_now = py + bvy_mean * snapshot/10.
pz_now = pz + bvz_mean * snapshot/10.

cM = [px_now*ppc, py_now*ppc, pz_now*ppc]
#rad = 60

le = [cM[0] - rad*ppc, cM[1]-rad*ppc, cM[2]-rad/2.0*ppc]
re = [cM[0] + rad*ppc, cM[1]+rad*ppc, cM[2]+rad/2.0*ppc]

# Generate regions to access the particle and fluid data.
box  = pf.region(cM, le, re)
sph  = pf.sphere(cM, rad*pc)

prj = pf.proj("numdens", "z", center=cM, data_source=box)

#resolution_frb = 512
resolution_frb = 1024

frb = prj.to_frb(rad*pc, resolution_frb, center=cM)["numdens"].value

# Make a Projection of the mass weighted Magnetic field in the plane of "sky"

resolution_frb = 1024

lic_res = resolution_frb / 2.

#prj_magx = pf.proj("magx", "z", center=cM, data_source=box, weight_field="cell_mass")
prj_magx = pf.proj("magx", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magx = prj_magx.to_frb(rad*pc, resolution_frb, center=cM)["magx"].value

#prj_magy = pf.proj("magy", "z", center=cM, data_source=box, weight_field="cell_mass")
prj_magy = pf.proj("magy", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magy = prj_magy.to_frb(rad*pc, resolution_frb, center=cM)["magy"].value

#prj_magy = pf.proj("magy", "z", center=cM, data_source=box, weight_field="cell_mass")
prj_magz = pf.proj("magz", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magz = prj_magz.to_frb(rad*pc, resolution_frb, center=cM)["magz"].value


#prj = pf.proj("numdens", "z", center=cM, data_source=box, weight_field="cell_mass")
prj_B = pf.proj("magnetic_field_strength", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_B = prj_B.to_frb(rad*pc, resolution_frb, center=cM)["magnetic_field_strength"].value

prj_magx2 = pf.proj("magx", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magx2 = prj_magx.to_frb(rad*pc, 512, center=cM)["magx"].value

# Make a Projection of the mass weighted Magnetic field in the plane of "sky"

lic_res = int(resolution_frb / 2.)

#prj_magx = pf.proj("magx", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magx_lic = prj_magx.to_frb(rad*pc, lic_res, center=cM)["magx"].value

#prj_magy = pf.proj("magy", "z", center=cM, data_source=box, weight_field="cell_mass")
frb_magy_lic = prj_magy.to_frb(rad*pc, lic_res, center=cM)["magy"].value

vectors = np.zeros((lic_res,lic_res, 2),dtype=np.float32)

vectors[...,0] = frb_magx_lic
vectors[...,1] = frb_magy_lic

texture = np.random.rand(lic_res,lic_res).astype(np.float32)

kernellen = 15
kernel = np.sin(np.arange(kernellen)*np.pi/kernellen)
kernel = kernel.astype(np.float32)

image = lic_internal.line_integral_convolution(vectors, texture, kernel)
```
